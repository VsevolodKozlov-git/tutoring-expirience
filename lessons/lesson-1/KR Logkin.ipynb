```python
import numpy as np
import matplotlib.pyplot as plt

# матрицы преобразования
matrix = np.array([[[0.01, -1], [1, 0.7]],
                   [[0.01, 1], [-1, 0.7]],
                  [[0.01, -1], [-1, 0.7]],
                   [[0.01, 1], [1, 0.7]],
                  [[-0.01, -1], [1, 0.7]],
                   [[-0.01, 1], [-1, 0.7]],
                  [[-0.01, -1], [-1, 0.7]],
                   [[0.01, 1], [1, 0.7]]])

# получаем исходные координаты
theta = np.linspace(0, 2*np.pi, 100)
a = np.cos(theta)
b = np.sin(theta)*np.cos(theta)

vec = np.array([a,b])

# рисуем первый график
figure, axes = plt.subplots()
axes.plot(vec[0], vec[1])

# преобразуем исходные координаты
for i in range(8):
    vec = matrix[i].dot(vec)

# рисуем второй график
axes.plot(vec[0], vec[1])

# отображаем график
axes.set_aspect(1)
plt.title('Вариант 1 - Кривая Лемниската Жеромо ')
plt.show()

# сохраняем график в файл
figure.savefig('example.png')
```
